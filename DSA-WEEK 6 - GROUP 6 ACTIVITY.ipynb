```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
data = pd.read_excel('Dataset_w6.xlsx')
data
```

# Find out how many countries are there in the dataset

```python
n = len(pd.unique(data['country']))
print("Number of Countries available in the given dataset is ", n)
```

# Examine the data by using visualizations

```python
data1=data[["Region", "Population in thousands (2017)"]].groupby("Region").sum()
data1=pd.DataFrame(data1.sort_values('Population in thousands (2017)', ascending=False))
```

```python
data1.plot(kind='bar',figsize=(16,12))
plt.title('Region vs Total Population(2017)', fontsize=22,fontweight="bold")
plt.xlabel('Region', fontsize=16,fontweight="bold")
plt.ylabel('Population', fontsize=16,fontweight="bold")
plt.xticks(fontsize=16,rotation=90)
plt.yticks(np.arange(0,2000000,200000))
plt.legend(loc=1,fontsize=16)
plt.grid()
plt.show()
```

Southern Asia having highest population with population greater than 1800 million

```python
data2=data[["Region", "GDP per capita (current US$)"]].groupby("Region").mean()
data2=pd.DataFrame(data2.sort_values('GDP per capita (current US$)', ascending=False))
```

```python
data2.plot(kind='bar',figsize=(16,12))
plt.title('Region vs Mean GDP Per Capita(USD)', fontsize=22,fontweight="bold")
plt.xlabel('Region', fontsize=16,fontweight="bold")
plt.ylabel('Mean GDP Per Capita(USD)', fontsize=16,fontweight="bold")
plt.xticks(fontsize=16,rotation=90)
plt.yticks(np.arange(0,60000,4000))
plt.legend(loc=1,fontsize=16)
plt.grid()
plt.show()
```

Northern America have mean highest Pecapita GDP and Western Africa have lowest

```python
data3=data[["Region", "Agricultural production index "]].groupby("Region").mean()
data3=pd.DataFrame(data3.sort_values('Agricultural production index ', ascending=False))
data3.plot(kind='bar',figsize=(16,12))
plt.title('Region vs Mean Agricultural Production Index', fontsize=22,fontweight="bold")
plt.xlabel('Region', fontsize=16,fontweight="bold")
plt.ylabel('Mean Agricultural production index', fontsize=16,fontweight="bold")
plt.xticks(fontsize=16,rotation=90)
plt.yticks()
plt.legend(loc=1,fontsize=16)
plt.grid()
plt.show()
```

South eastern Asia have highest Agricultural production index and Micron asia have lowest

```python
data4=data[["Region", "International trade: Imports (million US$)"]].groupby("Region").sum()
data4=pd.DataFrame(data4.sort_values('International trade: Imports (million US$)', ascending=False))
data4.plot(kind='bar',figsize=(16,12))
plt.title('Region vs Total International Trade: Imports (million US$)', fontsize=22,fontweight="bold")
plt.xlabel('Region', fontsize=16,fontweight="bold")
plt.ylabel('Total International trade-Imports(million US$)', fontsize=16,fontweight="bold")
plt.xticks(fontsize=16,rotation=90)
plt.yticks()
plt.legend(loc=1,fontsize=16)
plt.grid()
plt.show()
```

when considering Total international Trade import Eastern Asia is on Top1 position

# Find out which region has more consistent with respect to GDP per capita (current US$)

```python
df4=data[["Region", "GDP per capita (current US$)"]].groupby("Region").std()
df4=pd.DataFrame(df4.sort_values('GDP per capita (current US$)', ascending=False))
print ('The table below shows the standard deviation of of GDP per capita region wise ')
df4
```
